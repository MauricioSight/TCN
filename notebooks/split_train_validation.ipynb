```python
import torch
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
```

```python
cash_path = '../data/tow-ids-dataset/processed/train_tow-ids.pt'
cache = torch.load(cash_path, weights_only=False)
values, labels = cache['values'], cache['labels']
```

```python
labels['timestamp'] = labels['timestamp'].astype(float)
labels['timestamp'] = pd.to_datetime(labels['timestamp'], unit='s')
```

```python
freq = labels.set_index('timestamp').resample('1S').size().reset_index(name='freq')

fig = px.line(freq, x='timestamp', y='freq',
              title="Frequência de Pacotes por Segundo",
              labels={'timestamp': 'Tempo', 'freq': 'Número de Pacotes'})

fig.show() 
```
